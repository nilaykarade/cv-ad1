```python
import cv2,os,tensorflow as tf, numpy as np
from tensorflow.python.keras.utils.data_utils import get_file
```

```python
class_file="coco.names"
model_URL="http://download.tensorflow.org/models/object_detection/tf2/20200711/ssd_mobilenet_v2_320x320_coco17_tpu-8.tar.gz"
```

```python
with open(class_file,'r') as f:
    class_list=f.read().splitlines()
    
class_list
```

```python
color_list=np.random.uniform(0,255,(len(class_list)))
color_list
```

```python
len(class_list),len(color_list)
```

```python
model_file=os.path.basename(model_URL)
model_file
```

```python
model_name=model_file[:-7]
model_name
```

```python
models_dir="./pretrained_models"
os.makedirs(models_dir,exist_ok=True)
```

```python
#download model
get_file(fname=model_file,origin=model_URL,cache_dir=models_dir,cache_subdir="checkpoints",extract=True)
```

```python
tfod_model=tf.saved_model.load(os.path.join(models_dir,"checkpoints",model_name,"saved_model"))
print(str(model_name)+" loaded....")
```

```python
#Predict image class
image_path="test/5.jpg"
image=cv2.imread(image_path)
#image=cv2.resize(image,[400,600])
"""cv2.imshow("input image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()"""
```

```python
class_indexes=detections['detection_classes'][0].numpy().astype(np.int32)
class_indexes
```

```python
display_text=class_list[class_indexes[0]]
display_text
```

```python
class_color=color_list[class_indexes[0]]
```

```python
#get values of bbox
#below values are normalised x,y cordinates
ymin,xmin,ymax,xmax=bboxs[0]
#converting them to actual x,y values
xmin,xmax,ymin,ymax=(xmin*iW,xmax*iW,ymin*iH,ymax*iH)
#converting flaot to int
xmin,xmax,ymin,ymax=int(xmin),int(xmax),int(ymin),int(ymax)
```

```python
xmin,xmax,ymin,ymax
```

```python
cv2.rectangle(image,(xmin,ymin),(xmax,ymax), color=class_color,thickness=2)
cv2.putText(image,display_text,(20,20),cv2.FONT_HERSHEY_COMPLEX,1,class_color,1)
cv2.imwrite("p1.jpg",image)
```

```python
#bbox_idx=tf.image.non_max_supression(bboxs,class_scores,max_output_size=50,iou_treshold=0.5,score_threshold=0.5)

#if bbox is found i.e. object is detected
if len(bboxs)!=0:
    for i in range(0,len(bboxs)):
        bbox=tuple(bboxs[i].tolist())
        class_confidence=round(100*class_scores[i])
        class_index=class_indexes[i]
        
        if class_confidence>50.0:            
            print("class_confidence--",class_confidence)
            class_lbl_txt=class_list[class_index]
            class_color=color_list[class_index]

            display_text=str(class_lbl_txt)+" "+str(class_confidence)

            #get values of bbox
            #below values are normalised x,y cordinates
            ymin,xmin,ymax,xmax=bbox
            #converting them to actual x,y values
            xmin,xmax,ymin,ymax=(xmin*iW,xmax*iW,ymin*iH,ymax*iH)
            #converting flaot to int
            xmin,xmax,ymin,ymax=int(xmin),int(xmax),int(ymin),int(ymax)

            cv2.rectangle(image,(xmin,ymin),(xmax,ymax), color=class_color,thickness=2)
            cv2.putText(image,display_text,(xmin,ymin-10),cv2.FONT_HERSHEY_PLAIN,1,class_color,2)

    #cv2.imshow("result",image)
    cv2.imwrite("p5.jpg",image)
```
